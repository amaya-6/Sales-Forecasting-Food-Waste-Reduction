```python
import pandas as pd
```

```python
df = pd.read_csv("SuperMarket Analysis.csv")
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
```

```python
df["date"] = pd.to_datetime(df["date"])
df["time"] = pd.to_datetime(df["time"]).dt.time
```

```python
df["hour"] = pd.to_datetime(df["time"].astype(str)).dt.hour
```

```python
df.isnull().sum()
```

```python
df.duplicated(subset="invoice_id").sum()
df = df.drop_duplicates(subset="invoice_id")
```

```python
df[df["quantity"] <= 0]
df[df["unit_price"] <= 0]
df = df[(df["quantity"] > 0) & (df["unit_price"] > 0)]
```

```python
df["calculated_sales"] = df["unit_price"] * df["quantity"] + df["tax_5%"]
```

```python
(df["sales"] - df["calculated_sales"]).abs().describe()
```

```python
df["day"] = df["date"].dt.day
df["day_of_week"] = df["date"].dt.day_name()
df["month"] = df["date"].dt.month
df["week"] = df["date"].dt.isocalendar().week
df["is_weekend"] = df["day_of_week"].isin(["Saturday", "Sunday"]).astype(int)
```

```python
daily_sales = (
    df.groupby(["date", "product_line"])
      .agg(
          total_quantity=("quantity", "sum"),
          total_sales=("sales", "sum")
      )
      .reset_index()
)
```

```python
df.to_csv("cleaned_SuperMarket_Analysis.csv", index=False)
```
